```python
import matplotlib.pyplot as plt

from pandas import DataFrame, read_csv

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import numpy as np
```

```python
df = read_csv("C:/Users/Benny Wysong-Grass/Desktop/all-times.csv")
df = df.iloc[: , 1:]
df
```

```python
pca = PCA()
principals = pca.fit(df)
```

```python
def get_out_of_class(df, principals):
    n = principals.components_[0]
    data = np.zeros((df.shape[0],2))
    for i in range(df.shape[0]):
        p = df.iloc[i].values
        data[i] = [i, np.linalg.norm(p-n*np.dot(n, p))]
    return data
```

```python
def plot_pca(df, principals):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    n0 = principals.components_[0]
    n1 = principals.components_[1]
    data = np.zeros((2,df.shape[0]))
    for i in range(df.shape[0]):
        p = df.iloc[i].values
        data[0][i] = np.dot(n0, p)
        data[1][i] = np.dot(n1, p)
    
    plt.scatter(data[0], data[1])
    plt.show()
```

```python
plot_pca(df, principals)
```

```python
data = get_out_of_class(df, principals)
data.sort()
```
